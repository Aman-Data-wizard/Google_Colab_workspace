```python
ax = house['total_bedrooms'].plot(kind='density', figsize=(14,6)) # kde
ax.axvline(house['total_bedrooms'].mean(), color='red')
ax.axvline(house['total_bedrooms'].median(), color='green')
```

```python
ax = house['total_bedrooms'].plot(kind='hist', figsize=(14,6))
ax.set_ylabel('Number of houses')
ax.set_xlabel('dollars')
```
