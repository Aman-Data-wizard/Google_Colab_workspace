```python
def get_historic_price(symbol, days=7):
    try:
        # Calculate date range
        end_date = datetime.now().strftime('%Y-%m-%d')  # Today’s date
        start_date = (datetime.now() - timedelta(days=days)).strftime('%Y-%m-%d')  # Past date

        # Initialize scraper with the correct parameters
        scraper = CmcScraper(symbol, start_date, end_date)

        # Fetch historical data as a Pandas DataFrame
        df = scraper.get_dataframe()

        if df.empty:
            raise ValueError(f"No data found for {symbol} in the given date range!")

        return df

    except Exception as e:
        print(f"Error fetching historical data for {symbol}: {e}")
        return None  # Return None instead of crashing

# Fetch data for Bitcoin and Ethereum
bitcoin_data = get_historic_price("BTC", days=7)
ethereum_data = get_historic_price("ETH", days=7)

# Print data if available
if bitcoin_data is not None:
    print("Bitcoin Data (Last 7 Days):")
    print(bitcoin_data)

if ethereum_data is not None:
    print("\nEthereum Data (Last 7 Days):")
    print(ethereum_data)
```
